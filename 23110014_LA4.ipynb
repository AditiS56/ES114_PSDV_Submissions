```python
import matplotlib.pyplot as plt
import numpy as np
import math

## (a) UNIFORM


def uniform(a, b, x):
  if((x>=a)&(x<=b)):
    p=1/(b-a)
  elif(x<a):
    p=0
  else:
    p=0
  return p

def uniform_cdf(a, b, x):
  if((x>=a)&(x<=b)):
    c=(x-a)/(b-a)
  elif(x<a):
    c=0
  else:
    c=1
  return c


xrange=np.linspace(0, 28, 1000)

#parameter 1

a1=2
b1=6
pdf1=[uniform(a1,b1,x) for x in xrange]
cdf1=[uniform_cdf(a1,b1,x) for x in xrange]

#parameter 2

a2=5
b2=11
pdf2=[uniform(a2,b2,x) for x in xrange]
cdf2=[uniform_cdf(a2,b2,x) for x in xrange]

#parameter 3

a3=7
b3=25
pdf3=[uniform(a3,b3,x) for x in xrange]
cdf3=[uniform_cdf(a3,b3,x) for x in xrange]


#graph

fig, axes = plt.subplots(1,2,figsize=(15, 5))

# PDF

axes[0].plot(xrange, pdf1, color='gold', label='parameter 1')
axes[0].plot(xrange, pdf2, color='orange', label='parameter 2')
axes[0].plot(xrange, pdf3, color='red', label='parameter 3')
axes[0].fill_between(xrange, pdf1, color='gold', alpha=0.2)
axes[0].fill_between(xrange,pdf2, color='orange', alpha=0.2)
axes[0].fill_between(xrange, pdf3, color='red', alpha=0.2)
axes[0].legend(loc="upper right")
axes[0].grid(linestyle='--')
axes[0].set_title("UNIFORM PDF")
axes[0].set_xlabel('Events')
axes[0].set_ylabel("PDF")

# CDF

axes[1].plot(xrange, cdf1, color='gold', label='parameter 1')
axes[1].plot(xrange, cdf2, color='orange', label='parameter 2')
axes[1].plot(xrange, cdf3, color='red', label='parameter 3')
axes[1].fill_between(xrange, cdf1, color='gold', alpha=0.2)
axes[1].fill_between(xrange,cdf2, color='orange', alpha=0.2)
axes[1].fill_between(xrange, cdf3, color='red', alpha=0.2)
axes[1].legend(loc="lower right")
axes[1].grid(linestyle='--')
axes[1].set_title("UNIFORM CDF")
axes[1].set_xlabel('Events')
axes[1].set_ylabel("CDF")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math


## (d) CAUCHY

def cauchy(x, x0, gamma):
  p=1/(np.pi*gamma*(1+((x-x0)/gamma)**2))
  return p

def cauchy_cdf(x, x0, gamma):
  c=np.arctan((x-x0)/gamma)/np.pi+0.5
  return c

xrange=np.linspace(-25, 25, 1000)

x0=0

#parameter 1
gamma1=5
pdf1=[cauchy(x,x0,gamma1) for x in xrange]
cdf1=[cauchy_cdf(x,x0,gamma1) for x in xrange]

#parameter 2
gamma2=3
pdf2=[cauchy(x,x0,gamma2) for x in xrange]
cdf2=[cauchy_cdf(x,x0,gamma2) for x in xrange]

#parameter 3
gamma3=2
pdf3=[cauchy(x,x0,gamma3) for x in xrange]
cdf3=[cauchy_cdf(x,x0,gamma3) for x in xrange]

fig, axes = plt.subplots(1,2,figsize=(15, 5))

# PDF

axes[0].plot(xrange, pdf1, color='dodgerblue', label='parameter 1')
axes[0].plot(xrange, pdf2, color='turquoise', label='parameter 2')
axes[0].plot(xrange, pdf3, color='teal', label='parameter 3')
axes[0].legend(loc="upper right")
axes[0].grid(linestyle='--')
axes[0].set_title("CAUCHY PDF")
axes[0].set_xlabel('Events')
axes[0].set_ylabel("PDF")

# CDF

axes[1].plot(xrange, cdf1, color='dodgerblue', label='parameter 1')
axes[1].plot(xrange, cdf2, color='turquoise', label='parameter 2')
axes[1].plot(xrange, cdf3, color='teal', label='parameter 3')
axes[1].legend(loc="lower right")
axes[1].grid(linestyle='--')
axes[1].set_title("CAUCHY CDF")
axes[1].set_xlabel('Events')
axes[1].set_ylabel("CDF")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math


## (e) LAPLACIAN

def laplacian(x, b, mu):
  if(x>=mu):
    p=np.exp((mu-x)/b)/(2*b)
  else:
    p=np.exp((x-mu)/b)/(2*b)
  return p

def laplacian_cdf(x,b,mu):
  if(x>=mu):
    c=1-np.exp((mu-x)/b)/2
  else:
    c=np.exp((x-mu)/b)/2
  return c

xrange=np.linspace(-25, 25, 1000)
mu=0

#parameter 1

b1=5
pdf1=[laplacian(x,b1,mu) for x in xrange]
cdf1=[laplacian_cdf(x,b1,mu) for x in xrange]

#parameter 2

b2=3
pdf2=[laplacian(x,b2,mu) for x in xrange]
cdf1=[laplacian_cdf(x,b2,mu) for x in xrange]

#parameter 3

b3=2
pdf3=[laplacian(x,b3,mu) for x in xrange]
cdf1=[laplacian_cdf(x,b3,mu) for x in xrange]

# GRAPH

fig, axes = plt.subplots(1,2,figsize=(15, 5))

# PDF

axes[0].plot(xrange, pdf1, color='coral', label='parameter 1')
axes[0].plot(xrange, pdf2, color='maroon', label='parameter 2')
axes[0].plot(xrange, pdf3, color='red', label='parameter 3')
axes[0].legend(loc="upper right")
axes[0].grid(linestyle='--')
axes[0].set_title("LAPLACIAN PDF")
axes[0].set_xlabel('Events')
axes[0].set_ylabel("PDF")

# CDF

axes[1].plot(xrange, cdf1, color='coral', label='parameter 1')
axes[1].plot(xrange, cdf2, color='maroon', label='parameter 2')
axes[1].plot(xrange, cdf3, color='red', label='parameter 3')
axes[1].legend(loc="lower right")
axes[1].grid(linestyle='--')
axes[1].set_title("LAPLACIAN CDF")
axes[1].set_xlabel('Events')
axes[1].set_ylabel("CDF")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math


## (a) UNIFORM CDF COMPARISON

def uniform(a, b, x):
  p=1/(b-a)
  return p

def uniform_cdf(a, b, x):
  c=(x-a)/(b-a)
  return c


a=5
b=13

xrange=np.arange(a, b, 0.001)


# using cumsum
pdf=[uniform(a,b,x) for x in xrange]
cdf_cum=np.cumsum(pdf)/1000
plt.plot(xrange,cdf_cum,color='purple',linestyle="-",label='using cumsum')


# using closed form expression

cdf=[uniform_cdf(a,b,x) for x in xrange]
plt.plot(xrange,cdf,color='orange',linestyle="dashed",label='closed form expression')

plt.title("UNIFORM CDF COMPARISON")
plt.legend(loc="lower right")
plt.show()
```

Q3. Plot the CDF of each discrete random variable given in Lab Assignment 3 by varying parameters values taken as user input (No need to derive closed form expression in this case).

```python
import matplotlib.pyplot as plt
import numpy as np
import math

## (a) DISCRETE CDF

def discrete_cdf(xrange, x, c):
  a= xrange[0]
  b= xrange[-1]
  if x<a:
    y=0
  elif x>b:
    y=1
  elif int(x) in xrange:
    y= ((int(x)-a+1)/(b-a+1))
  else:
    y= c
  return y

x=np.linspace(-1, 30, 10000)
n= x.size

#parameter 1
xrange1= np.arange(0,13)
cdf1= np.zeros(x.shape)

#parameter 2
xrange2= np.arange(2,17)
cdf2= np.zeros(x.shape)

#parameter 3
xrange3=np.arange(4, 21)
cdf3= np.zeros(x.shape)

for i in range(n):
  cdf1[i]= discrete_cdf(xrange1, x[i], cdf1[i-1])
  cdf2[i]= discrete_cdf(xrange2, x[i], cdf2[i-1])
  cdf3[i]= discrete_cdf(xrange3, x[i], cdf3[i-1])


plt.plot(x, cdf1,color='red', label='parameter 1')
plt.plot(x, cdf2, color='orange', label='parameter 2')
plt.plot(x, cdf3, color='green', label='parameter 3')
plt.legend(loc="lower right")
plt.grid(linestyle='--')
plt.title("DISCRETE CDF")
plt.xlabel('EVENTS')
plt.ylabel("CDF")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

## (b) BERNOULLI CDF

def bernoulli_cdf(x, p):
  if(x<0):
    y=0
  elif(x>1):
    y=1
  else:
    x=int(x)
    y=p**x*(1-p)**(1-x)
  return(y)

xrange=np.linspace(-5, 6, 1000)


#parameter 1
p1=0.3
cdf1=[bernoulli_cdf(x, p1) for x in xrange]

#parameter 2
p2=0.5
cdf2=[bernoulli_cdf(x, p2) for x in xrange]

#parameter 3
p3=0.7
cdf3=[bernoulli_cdf(x, p3) for x in xrange]

#graph

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].plot(xrange, cdf1, color='red',label="parameter 1")
axes[0].set_title('Parameter 1')
axes[0].set_xlabel('events')
axes[0].set_ylabel('CDF')
axes[0].legend()
axes[0].grid(linestyle='--')

axes[1].plot(xrange,cdf2, color='orange',label="parameter 2")
axes[1].set_title('Parameter 2')
axes[1].set_xlabel('events')
axes[1].set_ylabel('CDF')
axes[1].legend()
axes[1].grid(linestyle='--')

axes[2].plot(xrange, cdf3, color='green',label="parameter 3")
axes[2].set_title('Parameter 3')
axes[2].set_xlabel('events')
axes[2].set_ylabel('CDF')
axes[2].legend()
axes[2].grid(linestyle='--')

fig.suptitle('BERNOULLI CDF')
plt.tight_layout()


plt.show()




```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

## (c) BINOMIAL CDF

def factorial(x):
  fact=1
  for i in range(1,x+1):
    fact=fact*i
  return fact

def combinations(n,r):
  return factorial(n) / (factorial(r) * factorial(n - r))

def binomial_cdf(n,p,k):
  if k<0:
    y=0
  elif k>=n:
    y=1
  else:
    x=int(k)
    y=0
    for i in range(x+1):
      y=y+ combinations(n,i)*p**i*(1-p)**(n-i)
  return y



m=12
xrange=np.linspace(-1, m+1, 1000)


#parameter 1
p1=0.3
cdf1=[binomial_cdf(m,p1,x) for x in xrange]

#parameter 2
p2=0.6
cdf2=[binomial_cdf(m,p2,x) for x in xrange]

#parameter 3
p3=0.9
cdf3=[binomial_cdf(m,p3,x) for x in xrange]


plt.plot(xrange, cdf1, color='red', label='parameter 1')
plt.plot(xrange, cdf2, color='orange', label='parameter 2')
plt.plot(xrange, cdf3, color='green', label='parameter 3')
plt.legend(loc="lower right")
plt.grid(linestyle='--')
plt.title("BINOMIAL CDF")
plt.xlabel('x successes')
plt.ylabel("CDF")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

## (d) POISSON CDF


def factorial(x):
  fact=1
  for i in range(1,x+1):
    fact=fact*i
  return fact

def poisson_cdf(n,k,p):
  lam=n*p
  if k<0:
    y=0
  else:
    x=int(k)
    y=0
    for i in range(x+1):
      y=y+ (((np.exp(-lam))*lam**i)/factorial(i))
  return y

m=1000
xrange=np.linspace(-1, 100, 5000)

#parameter 1
p1=0.001
cdf1=[poisson_cdf(m,x,p1) for x in xrange]

#parameter 2
p2=0.003
cdf2=[poisson_cdf(m,x,p2) for x in xrange]

#parameter 3
p3=0.02
cdf3=[poisson_cdf(m,x,p3) for x in xrange]


plt.plot(xrange, cdf1, color='red', label='parameter 1')
plt.plot(xrange, cdf2, color='orange', label='parameter 2')
plt.plot(xrange, cdf3, color='green', label='parameter 3')
plt.legend(loc="lower right")
plt.grid(linestyle='--')
plt.title("POISSON CDF")
plt.xlabel('x successes')
plt.ylabel("CDF")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

## (e) GEOMETRIC CDF

def geometric_cdf(x, p):
  if(x<0):
    y=0
  else:
    x=int(x)
    y=0
    for i in range (1, x+1):
      y+=p*(1-p)**(i-1)
  return(y)

xrange=np.linspace(-1, 40, 1000)

#parameter 1

p1=0.7
cdf1=[geometric_cdf(x,p1) for x in xrange]

#parameter 2

p2=0.3
cdf2=[geometric_cdf(x,p2) for x in xrange]

#parameter 3

p3=0.5
cdf3=[geometric_cdf(x,p3) for x in xrange]


plt.plot(xrange, cdf1, color='red', label='parameter 1')
plt.plot(xrange, cdf2, color='orange', label='parameter 2')
plt.plot(xrange, cdf3, color='green', label='parameter 3')
plt.legend(loc="lower right")
plt.grid(linestyle='--')
plt.title("GEOMETRIC CDF")
plt.xlabel('x-1 fails')
plt.ylabel("CDF")
plt.show()
```
