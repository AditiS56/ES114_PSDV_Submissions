Q1. Plot the Probability Mass Functions of the following Discrete Random Variables by varying the appropriate parameters taken as input. Compare the plots of the same PMF plotted with different parameters.

a) Discrete Uniform

b) Bernoulli

c) Binomial

d) Poisson

e) Geometric

```python
import matplotlib.pyplot as plt
import numpy as np

## (a) DISCRETE UNIFORM

def discrete_un(X):
  first=X[0]
  last=X[-1]
  return 1/last-first+1

#parameter 1

xrange1=np.arange(8)
prob1=discrete_un(xrange1)

#parameter 2

xrange2=np.arange(5)
prob2=discrete_un(xrange2)

#parameter 3

xrange3=np.arange(13)
prob3=discrete_un(xrange3)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(xrange1, prob1, color='purple',edgecolor="orange",label="uniform")
axes[0].set_title('Parameter 1')
axes[0].set_ylim(0,2.5)
axes[0].legend()

axes[1].bar(xrange2, prob2,color='green',edgecolor="orange",label="uniform")
axes[1].set_title('Parameter 2')
axes[1].set_ylim(0,2.5)
axes[1].legend()

axes[2].bar(xrange3, prob3, color='skyblue',edgecolor="yellow",label="uniform")
axes[2].set_title('Parameter 3')
axes[2].set_ylim(0,2.5)
axes[2].legend()

for ax in axes:
    ax.set_xlabel('Events')
    ax.set_ylabel('Probability')

fig.suptitle('DISCRETE UNIFORM DISTRIBUTION')
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import numpy as np

## (b) BERNOULLI'S RANDOM VARIABLE

def bernoulli(p,k):
  if k==0:
    return 1-p
  elif k==1:
    return p

krange=np.array([0,1])
#parameter 1

p1=0.82
krange1=np.array(["0","1"])
prob1=[bernoulli(p1,k) for k in krange]

#parameter 2

p2=0.35
krange2=np.array(["fail","success"])
prob2=[bernoulli(p2,k) for k in krange]

#parameter 3

p3=0.5
krange3=np.array(["heads","tails"])
prob3=[bernoulli(p3,k) for k in krange]


fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(krange1, prob1, width=0.25, color='orange',edgecolor="red",label="bernoulli")
axes[0].set_title('Parameter 1')
axes[0].legend()

axes[1].bar(krange2, prob2, width=0.25,color='darkgreen',edgecolor="green",label="bernoulli")
axes[1].set_title('Parameter 2')
axes[1].legend()

axes[2].bar(krange3, prob3,width=0.25, color='skyblue',edgecolor="blue",label="bernoulli")
axes[2].set_title('Parameter 3')
axes[2].legend(loc="upper center")

for ax in axes:
    ax.set_xlabel('Trials')
    ax.set_ylabel('Probability')

fig.suptitle('BERNOULLI RANDOM VARIABLE')
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

## (c) BINOMIAL RANDOM VARIABLE

def factorial(x):
  f=1
  for i in range(2, x+1):
    f*=i
  return(f)

def comb(n, k):
  c=1
  if(k<(n-k)):
    for i in range(n, n-k, -1):
      c*=i
    c=c//factorial(k)
  else:
    for i in range(n, k, -1):
      c*=i
    c=c//factorial(n-k)
  return(c)

def binomial(p,m,k):
  return (comb(m,k)*(p**k)*((1-p)**(m-k)))

#parameter 1
p1=0.01
m1=20
krange1=np.arange(0,m1+1)
prob1=[binomial(p1,m1,k1) for k1 in krange1]

#parameter 2
p2=0.06
m2=10
krange2=np.arange(0,m2+1)
prob2=[binomial(p2,m2,k2) for k2 in krange2]

#parameter 3
p3=0.1
m3=18
krange3=np.arange(0,m3+1)
prob3=[binomial(p3,m3,k3) for k3 in krange3]


fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(krange1, prob1, color='maroon',edgecolor="pink",label="binomial")
axes[0].set_title('Parameter 1')
axes[0].legend()

axes[1].bar(krange2, prob2, color='darkgreen',edgecolor="orange",label="binomial")
axes[1].set_title('Parameter 2')
axes[1].legend()

axes[2].bar(krange3, prob3, color='blue',edgecolor="yellow",label="binomial")
axes[2].set_title('Parameter 3')
axes[2].legend()

for ax in axes:
    ax.set_xlabel('K successes in M trials')
    ax.set_ylabel('Probability')

fig.suptitle('BINOMIAL RANDOM VARIABLE')
plt.tight_layout()


plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

## (d) POISSON DISTRIBUTION

def factorial(x):
  f=1
  for i in range(2, x+1):
    f*=i
  return(f)

def exp(x):
  e=1
  value=1
  for i in range(1, 500):
    value*=x/i
    e+=value
  return(e)

def poisson(lam,k):
  y = (exp(-lam))*(lam**k)/factorial(k)
  return y

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

#parameter 1

m1=100
p1=0.01
lam1=m1*p1

krange1=np.arange(0,m1+1)

prob1=[poisson(lam1,k1) for k1 in krange1]


#parameter 2

m2=90
p2=0.03
lam2=m2*p2

krange2=np.arange(0,m2+1)

prob2=[poisson(lam2,k2) for k2 in krange2]


#parameter 3

m3=100
p3=0.1
lam3=m3*p3

krange3=np.arange(0,m3+1)

prob3=[poisson(lam3,k3) for k3 in krange3]


axes[0].bar(krange1, prob1, color='blue',edgecolor="skyblue",label="poisson")
axes[0].set_title('Parameter 1')
axes[0].set_xlabel('K successes in M independant trials')
axes[0].set_ylabel('Probability')
axes[0].legend()

axes[1].bar(krange2, prob2, color='red',edgecolor="orange",label="poisson")
axes[1].set_title('Parameter 2')
axes[1].set_xlabel('K successes in M independant trials')
axes[1].set_ylabel('Probability')
axes[1].legend()

axes[2].bar(krange3, prob3, color='green',edgecolor="yellow",label="poisson")
axes[2].set_title('Parameter 3')
axes[2].set_xlabel('K successes in M independant trials')
axes[2].set_ylabel('Probability')
axes[2].legend()

fig.suptitle('POISSON RANDOM VARIABLE')
plt.tight_layout()


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

## (e) GEOMETRIC RANDOM VARIABLE

def geometric(p,k):
  return ((1-p)**(k-1))*p

#parameter 1
p1=0.08
krange1=np.arange(0,20)
prob1=[geometric(p1,k1) for k1 in krange1]

#parameter 2
p2=0.04
krange2=np.arange(0,15)
prob2=[geometric(p2,k2) for k2 in krange2]

#parameter 3
p3=0.01
krange3=np.arange(0,30)
prob3=[geometric(p3,k3) for k3 in krange3]


fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].bar(krange1, prob1, color='skyblue',edgecolor="blue",label="geometric")
axes[0].set_title('Parameter 1')
axes[0].set_xlabel('K independant trials')
axes[0].set_ylabel('Probability')
axes[0].legend()

axes[1].bar(krange2, prob2, color='yellow',edgecolor="pink",label="geometric")
axes[1].set_title('Parameter 2')
axes[1].set_xlabel('K independant trials')
axes[1].set_ylabel('Probability')
axes[1].legend()

axes[2].bar(krange3, prob3, color='orange',edgecolor="red",label="geometric")
axes[2].set_title('Parameter 3')
axes[2].set_xlabel('K independant trials')
axes[2].set_ylabel('Probability')
axes[2].legend()

fig.suptitle('GEOMETRIC RANDOM VARIABLE')
plt.tight_layout()


plt.show()

```

Q2. Show the equivalence of Binomial and Poisson for large values of M and small values of p.



```python
### EQUIVALENCE OF POISSON AND BINOMIAL


import matplotlib.pyplot as plt
import numpy as np
import math

##BINOMIAL

def factorial(x):
  f=1
  for i in range(2, x+1):
    f*=i
  return(f)

def comb(n, k):
  c=1
  if(k<(n-k)):
    for i in range(n, n-k, -1):
      c*=i
    c=c//factorial(k)
  else:
    for i in range(n, k, -1):
      c*=i
    c=c//factorial(n-k)
  return(c)

def binomial(p,m,k):
  return (comb(m,k)*(p**k)*((1-p)**(m-k)))

##POISSON

def factorial(x):
  f=1
  for i in range(2, x+1):
    f*=i
  return(f)

def exp(x):
  e=1
  value=1
  for i in range(1, 500):
    value*=x/i
    e+=value
  return(e)

def poisson(lam,k):
  y = (exp(-lam))*(lam**k)/factorial(k)
  return y



m=10000
p=0.001
l=m*p
krange=np.arange(0,30)
lam =float(l)


prob_pois=[poisson(lam,k) for k in krange]
prob_bin=[binomial(p,m,k) for k in krange]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
plt.legend(loc='upper right')
bar_width =0.40
index = np.arange(len(krange))


#method 1

axes[0].bar(index - bar_width/2, prob_pois, bar_width, color='purple', label='POISSON')
axes[0].bar(index + bar_width/2, prob_bin, bar_width, color='orange', label='BINOMIAL')
axes[0].set_title('Method 1')
axes[0].set_xlabel('K successes in M independant trials')
axes[0].set_ylabel('Probability')
axes[0].legend(loc='upper right')



#method 2

axes[1].bar(krange, prob_pois, color='skyblue',label='POISSON')
axes[1].plot(krange,prob_bin,color="blue",label="BINOMIAL")
axes[1].set_title('Method 2')
axes[1].set_xlabel('K successes in M independant trials')
axes[1].set_ylabel('Probability')
axes[1].legend(loc='upper right')

#method 3

axes[2].plot(krange, prob_pois, color='black',linestyle="-",label='POISSON')
axes[2].plot(krange,prob_bin,color="yellow",linestyle="dashed",label="BINOMIAL")
axes[2].set_title('Method 3')
axes[2].set_xlabel('K successes in M independant trials')
axes[2].set_ylabel('Probability')
axes[2].legend(loc='upper right')


fig.suptitle('EQUIVALENCE OF BINOMIAL AND POISSON')

plt.tight_layout()

plt.show()
```
