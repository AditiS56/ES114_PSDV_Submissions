```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##1

def mean(arr):
  sum=0
  mean=0
  for i in arr:
    sum+=i
  mean=sum/len(arr)
  return mean

def variance(arr):
  s=0
  var=0
  for i in arr:
    s+=(i-mean(arr))**2
    var=s/len(arr)
  return var
```

```python
data = sns.load_dataset(name='iris')
data
```

```python
sl=data['sepal_length'].values

print("Mean of Sepal Length : ",mean(sl))
print("Variance of Sepal Length : ",variance(sl))
```

```python
sw=data['sepal_width'].values

print("Mean of Sepal Width : ",mean(sw))
print("Variance of Sepal Width : ",variance(sw))
```

```python
pl=data['petal_length'].values

print("Mean of Petal Length : ",mean(pl))
print("Variance of Petal Length : ",variance(pl))
```

```python
pw=data['petal_width'].values

print("Mean of Petal Width : ",mean(pw))
print("Variance of Petal Width : ",variance(pw))
```

```python
def covariance(x,y):
  ex=mean(x)
  ey=mean(y)

  num=sum((x[i]-ex)*(y[i]-ey) for i in range(len(x)))
  cov=num/len(x)
  return cov

n=data.shape[0]

def correlation_coeff(x,y):
  cov=covariance(x,y)
  v1=variance(x)
  v2=variance(y)
  cc=cov/(np.sqrt(v1)*np.sqrt(v2))
  return cc
```

```python
##2

cc=[]
columns= data.columns

for x in columns[:-1]:
  for y in columns[:-1]:
    if x!=y:
      cor_coeff= correlation_coeff((data[x].to_numpy()), data[y].to_numpy())
      if (x,y,cor_coeff) and (y,x,cor_coeff) not in cc:
        cc.append((x,y,cor_coeff))

for coeff in sorted(cc,key= lambda x: x[2], reverse = True):
    print(f"{coeff[0]} and {coeff[1]}: {coeff[2]}")
```

```python
D=data.iloc[:, :4].to_numpy()

plt.figure(figsize=(15, 3))
plt.subplot(1, 3, 1)
plt.scatter(D[:, 0], D[:, 1], c='orange', s=60, alpha=0.4)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.subplot(1, 3, 2)
plt.scatter(D[:, 0], D[:, 2], c='purple', s=60, alpha=0.4)
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.subplot(1, 3, 3)
plt.scatter(D[:, 0], D[:, 3], c='green', s=60, alpha=0.4)
plt.xlabel('sepal_length')
plt.ylabel('petal_width')
plt.suptitle("CORRELATION BETWEEN ATTRIBUTES",fontsize=14)
plt.show()

print()

plt.figure(figsize=(15, 3))
plt.subplot(1, 3, 1)
plt.scatter(D[:, 1], D[:, 3], c='blue', s=60, alpha=0.4)
plt.xlabel('sepal_width')
plt.ylabel('petal_width')
plt.subplot(1, 3, 2)
plt.scatter(D[:, 1], D[:, 2], c='pink', s=60, alpha=0.4)
plt.xlabel('sepal_width')
plt.ylabel('petal_length')
plt.subplot(1, 3, 3)
plt.scatter(D[:, 2], D[:, 3], c='red', s=60, alpha=0.4)
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```
